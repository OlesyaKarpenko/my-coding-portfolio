```python
#!/usr/bin/env python
# coding: utf-8

# In[61]:


import numpy as np
import math
import matplotlib.pyplot as plt

n0 = 4
n1 = 8
n2 = 16

a = 0
b = 1

x0 = np.arange(a, b + (b - a)/n0, (b - a)/n0)
x1 = np.arange(a, b + (b - a)/n1, (b - a)/n1)
x2 = np.arange(a, b + (b - a)/n2, (b - a)/n2)
yreal = b * math.acos(b) - math.sqrt(1 - b**2) - a * math.acos(a) + math.sqrt(1 - a**2)

print ("Узлы при n0:\n", x0)
print ("Узлы при n1:\n", x1)
print ("Узлы при n2:\n", x2)
print ("\nРеальное значение интеграла: ", yreal)


# In[62]:


def f(x):
    return math.acos(x)


# In[63]:


#for n0:
y0_trapez = (f(x0[0]) + f(x0[n0])) / 2

for i in range(1, n0):
    y0_trapez = y0_trapez + f(x0[i])

y0_trapez = y0_trapez / n0


#for n1:
y1_trapez = (f(x1[0]) + f(x1[n1])) / 2

for i in range(1, n1):
    y1_trapez = y1_trapez + f(x1[i])

y1_trapez = y1_trapez / n1


#for n2:
y2_trapez = (f(x2[0]) + f(x2[n2])) / 2

for i in range(1, n2):
    y2_trapez = y2_trapez + f(x2[i])

y2_trapez = y2_trapez / n2


# In[64]:


def f_gauss(a1, b1):
    x_1 = (a1 + b1) / 2 + math.sqrt(3) * (b1 - a1) / 6
    x_2 = (a1 + b1) / 2 - math.sqrt(3) * (b1 - a1) / 6
    
    return (b1 - a1) * (f(x_1) +f(x_2)) / 2


# In[65]:


#for n0
y0_gauss = 0

for i in range(1, n0 + 1):
    y0_gauss = y0_gauss + f_gauss(x0[i - 1], x0[i])
    
#for n1
y1_gauss = 0

for i in range(1, n1 + 1):
    y1_gauss = y1_gauss + f_gauss(x1[i - 1], x1[i])
    
    
#for n2
y2_gauss = 0

for i in range(1, n2 + 1):
    y2_gauss = y2_gauss + f_gauss(x2[i - 1], x2[i])


# In[66]:


print ("\nЗначение интеграла (реальное): ", yreal)
print("\n")
print ("\nЗначение интеграла (трапецией) для n0: ", y0_trapez)
print ("\nЗначение интеграла (трапецией) для n1: ", y1_trapez)
print ("\nЗначение интеграла (трапецией) для n2: ", y2_trapez)
print("\n")
print ("\nЗначение интеграла (Гаусс с 2 узлами) для n0: ", y0_gauss)
print ("\nЗначение интеграла (Гаусс с 2 узлами) для n1: ", y1_gauss)
print ("\nЗначение интеграла (Гаусс с 2 узлами) для n2: ", y2_gauss)


# In[67]:


#рассчет погрешности для трапеции


# In[68]:


print ("\nТеоретическое значение: 4")

print ("\nЗначение для n0/n1: ", abs (1 - y0_trapez) / abs (1 - y1_trapez))
print ("\nЗначение для n1/n2: ", abs (1 - y1_trapez) / abs (1 - y2_trapez))


# In[69]:


#рассчет погрешности для Гаусса с двумя узлами


# In[70]:


print ("\nТеоретическое значение: 16")

print ("\nЗначение для n0/n1: ", abs (1 - y0_gauss) / abs (1 - y1_gauss))
print ("\nЗначение для n1/n2: ", abs (1 - y1_gauss) / abs (1 - y2_gauss))


# In[71]:


#найдем p и C для трапеции


# In[72]:


p_trapez = math.log(((y1_trapez - y0_trapez) / (y2_trapez - y1_trapez)), 2)
print("Значение p для трапеции: ", p_trapez)


# In[73]:


c_trapez = (y1_trapez - y0_trapez) / (((b - a) / n0) ** p_trapez *(1 - (1 / 2) ** p_trapez))
print("Значение C для трапеции: ", c_trapez)


# In[74]:


#найдем p и C для Гаусса с двумя узлами


# In[75]:


p_gauss = math.log(((y1_gauss - y0_gauss) / (y2_gauss - y1_gauss)), 2)
print("Значение p для Гаусса с двумя узлами: ", p_gauss)


# In[76]:


c_gauss = (y1_gauss - y0_gauss) / (((b - a) / n0) ** p_gauss *(1 - (1 / 2) ** p_gauss))
print("Значение C для Гаусса с двумя узлами: ", c_gauss)

```
